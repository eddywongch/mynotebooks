```python
def parseEdgeId(str):
        result = {}

        str_list = str.split('/')
        edge_str = str_list[1]
        edge_list = edge_str.split('-')
        
        result['source_label'] = edge_list[0]
        result['dest_label'] = edge_list[2]
        result['edge_label'] = edge_list[1]
        result['source_id'] = str_list[2]
        result['dest_id'] = str_list[3]

        return result
```

```python
s = parseEdgeId('dseg:/person-implied-keyword/bf720fd3069c1704fd25359e1ebb77e531c4949a/break')
print(s)
```

```python
import networkx as nx
GG = nx.Graph()

edges = g.E()
for e in edges:
    print("***", e.id )
    es = parseEdgeId(e.id)
    GG.add_edge(es['source_id'], es['dest_id'], edge_label=es['edge_label'], \
                source_label=es['source_label'], dest_label=es['dest_label'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
nx.draw(GG)
```

```python
dh = nx.degree_histogram(GG)
low = min(d for n, d in GG.degree())

print("degree (#nodes) ****")
for i in range(low, len(dh)):
    bar=''.join(dh[i]*['*'])
    print ('{:4}'.format(i), '{:4}'.format(dh[i]), bar)
```

```python
for v in GG.nodes():   # Some properties of the graph
    print (v, GG.degree(v) , nx.clustering(GG,v) )
```

```python
print("radius: %d" % nx.radius(GG))
print("diameter: %d" % nx.diameter(GG))
print("eccentricity: %s" % nx.eccentricity(GG))
print("center: %s" % nx.center(GG))
print("periphery: %s" % nx.periphery(GG))
print("density: %s" % nx.density(GG))
```
